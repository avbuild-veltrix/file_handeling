```python
#Handeling multiple specific exception in separate blocks
try:
    10/0
except ZeroDivisionError as e:
    print("The devsion is not possible because of the error,", e)
except TypeError as e:
    print("This is value error")
```

```python
import logging
logging.basicConfig(filename = "program.log", level= logging.DEBUG)
try:
    file = open("exe.txt","r")
    file.read()
except FileNotFoundError as e:
    print(f"The following file is not created and the error occured is,{e}")
    logging.error(f"The following file is not created and the error occured is,{e}")
```

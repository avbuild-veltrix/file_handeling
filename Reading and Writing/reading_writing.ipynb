```python
#Making an csv(COMMA SEPARATED VALUES FILE)
data = [["Name","Course","Fee"],
        ["Ajay","DS","20000"],
        ["Kartik","CN","5000"]]
```

```python
import csv
```

```python
with open("file_csv.csv","w") as f:
    w = csv.writer(f)
    for i in data:
        w.writerow(i)
```

```python
with open("file_csv.csv","r") as f:
    r = csv.reader(f)
    for i in r:
        print(i)
```

```python
#Buffered reading and writing
import io
```

```python
with open("buff.txt","wb") as f:
    file = io.BufferedWriter(f)
    file.write(b"The representation of your hex-encoded string is Good morning kartik.Hexadecimal BreakdownEach sequence represents a single character encoded in hexadecimal (base-16) ASCII. Here is how every individual byte translates into plain text:")
    file.write(b"I'm Kartik")
    file.write(b"Good morning kartik")
    file.flush()
```

```python
with open("buff.txt","rb") as f:
    file = io.BufferedReader(f)
    data = file.read(100)
    print(data)
```

```python
#USE CASES OF ELSE STATEMENT
try:
    file = open("file.txt", "w")
    file.write("Hello to exceptional handeling")#You will not be able to see the content of the file cause you didn't closed it yet
except Exception as e:
    print("There is an error",e)
    print(2+3)
else:
    print(3+54+23+23423)
    print("Kartik")
```

```python
#USE CASES OF ELSE STATEMENT
try:
    file = open("file.txt", "w")
    file.write("Hello to exceptional handeling")
except Exception as e:
    print("There is an error",e)
    print(2+3)
else:
    file.close()
    print(3+54+23+23423)
    print("Kartik")
```

```python
#FINALLY STATEMENT
try:
    file = open("example.txt", "w")
    file.write("Hello to exceptional handeling")
except Exception as e:
    print("There is an error",e)
    print(2+3)
else:
    file.close()
    print(3+54+23+23423)
    print("Kartik")
finally:
    print("This will be always executed")
```

```python
class ValidateSalary(Exception):
    def __init__(self, msg):
        self.msg = msg
```

```python
def validate_salary(salary):
    if salary <= 0:
        raise ValidateSalary("Negative salary is not possible")
    elif salary >= 30000000:
        raise ValidateSalary("Not reasonable")
    else:
        raise ValidateSalary("Valid salary")
```

```python
try:
    salary = int(input("Enter your salary"))
    validate_salary(salary)
except ValidateSalary as e:
    print(e)
```

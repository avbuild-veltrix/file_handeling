```python
l = [1, "Hello",[2,"World"],3,[4,"Python"]]

l1_int = []
l2_str = []

for i in l:
    logging.info(f"processing sublist element:{i}")
    
    if type(i) == list:
        for j in i:
            logging.info(f"processing sublist element:{j}")
            if type(j) == int:
                l1_int.append(j)
    elif type(i) == int:
            l1_int.append(i)
    else:
       l2_str.append(i)
        
logging.info(f"The result is : {l1_int,l2_str}")
logging.shutdown()
```

```python
#Now we will make multiple thread to execute the same program to reduce the time of execution of the code
#Since the ran sequentially (one thread on single processor), that's why its takes n*t(n is the number of funtion called) sec time for executing the program
import time
import threading

start = time.perf_counter()

def fun_test():
    print("Do something")
    print("Sleep for seconds")
    time.sleep(1)
    print("Done with sleeping")

threads = []

for i in range(10):
    t = threading.Thread(target = fun_test)
    t.start()
    threads.append(t)
    
for thread in threads:
    t.join()

end = time.perf_counter()
print("Total time ",(end-start))
```

```python
import time
import concurrent.futures

start = time.perf_counter()

url_list = [
    'https://raw.githubusercontent.com/dscape/spell/master/test/resources/big.txt',
    'https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-no-swears.txt',
    'https://raw.githubusercontent.com/itsfoss/text-files/master/sherlock.txt' ,
    'https://raw.githubusercontent.com/itsfoss/text-files/master/sample_log_file.txt',
]

data_list = ["data1a.txt","data2a.txt","data3a.txt","data4a.txt"]

import urllib.request

def download_data(url, filename):
    urllib.request.urlretrieve(url,filename)

    with concurrent.futures.ThreadPoolExecutor() as executor:
        executor.map(download_data,url_list,data_list)


end = time.perf_counter()
print("Total time ",(end-start))
                         
                        
```

```python
#Shared variable across all the threads
import time
import threading
import concurrent.futures

start = time.perf_counter()
shared_counter = 0
counter_lock = threading.Lock()

def increasing_shared_counter(x):
    global shared_counter
    with counter_lock:
        shared_counter = shared_counter + 1
        print(f"Thread {x}: increases the shared counter to {shared_counter}")
        time.sleep(1)

threads = [threading.Thread(target = increasing_shared_counter, args = (i,)) for i in [1,2,3,4,5,6]]

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()

end = time.perf_counter()
print("Total time ",(end-start))
                        
```

```python
#Shared variable across all the threads
import time
import threading
import concurrent.futures

start = time.perf_counter()
shared_counter = 0
counter_lock = threading.Lock()

def increasing_shared_counter(x):
    global shared_counter
    with counter_lock:
        shared_counter = shared_counter + 1
        print(f"Thread {x}: increases the shared counter to {shared_counter}")
        time.sleep(1)

with concurrent.futures.ThreadPoolExecutor() as executor:
    thread_args = [1,2,3,4,5,6]
    executor.map(increasing_shared_counter, thread_args)

end = time.perf_counter()
print("Total time ",(end-start))
                        
```

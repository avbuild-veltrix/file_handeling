```python
import time
import threading
import concurrent.futures

start = time.perf_counter()
shared_counter = 0
counter_lock = threading.Lock()

def incrementing_shared_counter(x):
    global shared_counter
    with counter_lock:
        shared_counter += 1
        print(f"Thread {x}: increased the shared counter to {shared_counter}")

threads= [threading.Thread(target = incrementing_shared_counter, args = (i,)) for i in [1,2,3,4,5,6]]

for thread in threads:
    thread.start()
for thread in threads:
    thread.join()

end = time.perf_counter()
print(f"The program finised in {end-start} seconds")
```

```python
import multiprocessing
import time

start = time.perf_counter()
def test_fun():
    print("Do something")
    print("Sleep for 1 second")
    time.sleep(1)
    print("Done sleeping")

p1 = multiprocessing.Process(target = test_fun)
p2 = multiprocessing.Process(target = test_fun)
p3 = multiprocessing.Process(target = test_fun)
p4 = multiprocessing.Process(target = test_fun)

p1.start()
p2.start()
p3.start()
p4.start()

p1.join()
p2.join()
p3.join()
p4.join()

end = time.perf_counter()
print(f"The program finsihed in {round(end-start, 2)} seconds")
```

```python
import multiprocessing
import time

start = time.perf_counter()
def test_fun():
    print("Do something")
    print("Sleep for 1 second")
    time.sleep(1)
    print("Done sleeping")

x = int(input("Enter the number of processes :"))

processes = []
for i in range(x):
    p = multiprocessing.Process(target = test_fun)
    p.start()
    processes.append(p)

for process in processes:
    p.join()

end = time.perf_counter()
print(f"The program finsihed in {round(end-start, 2)} seconds")
```

```python
import multiprocessing
import time

def test_fun():
    print("Do something")
    print("Sleep for 1 second")
    time.sleep(1)
    print("Done sleeping")

if __name__ == "__main__":
    start = time.perf_counter()

    processes = []

    for i in range(4):
        p = multiprocessing.Process(target=test_fun)
        p.start()
        processes.append(p)

    for process in processes:
        process.join()

    end = time.perf_counter()
    print(f"The program finished in {end-start:.2f} seconds")
```
